```python
import pandas as pd

cam_df = pd.read_excel("/content/datargb.xlsx")
gsr_df = pd.read_excel("/content/datawcam.xlsx")

cam_df.head(), gsr_df.head()
```

```python
merged_df = pd.merge(cam_df, gsr_df, on="S. No.", how="inner", suffixes=("_cam", "_gsr"))
print("Merged shape:", merged_df.shape)
merged_df.head()
```

```python
final_df=merged_df
final_df.to_csv("/content/final_multisensor_dataset.csv", index=False)
```

```python
import numpy as np

def combine_health_status(row):
    if row['Liver Healthy'] == 'No' or row['Jaundice'] == 'No':
        return 'No'
    elif row['Liver Healthy'] == 'Mid' or row['Jaundice'] == 'Mid':
        return 'Mid'
    else:
        return 'Yes'

final_df['Health_Status'] = final_df.apply(combine_health_status, axis=1)

# Drop the original 'Liver Healthy' and 'Jaundice' columns
final_df = final_df.drop(columns=['Liver Healthy', 'Jaundice'])

# Drop 'Age_gsr' and 'Gender_gsr' columns
final_df = final_df.drop(columns=['Age_gsr', 'Gender_gsr'])

print("Final DataFrame columns:", final_df.columns)
display(final_df.head())
```

```python
final_df.to_csv("/content/final_multisensor_dataset.csv", index=False)
```
